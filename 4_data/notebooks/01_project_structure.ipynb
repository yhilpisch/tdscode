### Notebook 01 · Project Structure

This notebook turns the project layout chapter into something concrete: a starter tree, a visible root, and a short explanation of where files belong.

### Why this notebook exists
A clean folder structure is the first reliability habit. It makes the rest of the book easier to follow because data, code, and documentation stay in separate places.

```python
from pathlib import Path
import sys
import tempfile

cwd = Path.cwd().resolve()
BOOK_ROOT = cwd
for candidate in [cwd, cwd.parent, cwd / "4_data"]:
    if (candidate / "src").exists() or (candidate / "code").exists():
        BOOK_ROOT = candidate
        break
CODE_DIR = BOOK_ROOT / "code"
if str(CODE_DIR) not in sys.path:
    sys.path.insert(0, str(CODE_DIR))

from project_scaffold import scaffold

print("Book root:", BOOK_ROOT)
```

```python
tmpdir = Path(tempfile.mkdtemp())
starter_root = scaffold(tmpdir / "starter_project")
files = sorted(str(path.relative_to(starter_root)) for path in starter_root.rglob("*") if path.is_file())
print("\n".join(files))
```

```python
project_parts = {
    "data": BOOK_ROOT / "data" / "capstone_project_health.csv",
    "src": BOOK_ROOT / "src" / "data_checks.py",
    "tests": BOOK_ROOT / "tests" / "test_data_checks.py",
    "notebooks": BOOK_ROOT / "notebooks" / "01_project_structure.ipynb",
}
lines = [f"{name:10s} -> {rel_path.relative_to(BOOK_ROOT)} -> exists={rel_path.exists()}" for name, rel_path in project_parts.items()]
print("\n".join(lines))
```
