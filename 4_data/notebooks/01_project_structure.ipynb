### Why this notebook exists
A clean folder structure is the first reliability habit. It makes the rest of the
book easier to follow because data, code, and documentation stay in separate places.

## Setup
We start by fixing the book root and import path so the scaffold helper and the
later file checks run from a predictable location.


```python
from pathlib import Path
import os
import subprocess
import sys

BOOK_NAME = "4_data"
REPO_URL = "https://github.com/yhilpisch/tdscode"

cwd = Path.cwd().resolve()
BOOK_ROOT = None
for candidate in (cwd, *cwd.parents):
    if candidate.name == BOOK_NAME and (candidate / "notebooks").exists():
        BOOK_ROOT = candidate
        break
    if (candidate / BOOK_NAME / "notebooks").exists():
        BOOK_ROOT = candidate / BOOK_NAME
        break

if BOOK_ROOT is None:
    repo_root = Path("/content/tdscode")
    if not repo_root.exists():
        # Clone the companion repo once in Colab.
        subprocess.run(
            ["git", "clone", "--depth", "1", REPO_URL, str(repo_root)],
            check=True,
        )
    BOOK_ROOT = repo_root / BOOK_NAME

os.chdir(BOOK_ROOT)

# Make the book root and the helper folder importable.
for path in (BOOK_ROOT, BOOK_ROOT / "code"):
    if path.exists() and str(path) not in sys.path:
        sys.path.insert(0, str(path))

requirements = BOOK_ROOT / "requirements.txt"
if requirements.exists() and "google.colab" in sys.modules:
    # Keep Colab aligned with the book dependencies.
    subprocess.run(
        [sys.executable, "-m", "pip", "install", "-q", "-r", str(requirements)],
        check=True,
    )

print("Book root:", BOOK_ROOT)
```

## Load the Scaffold Helper
The first project-specific code cell imports the helper that knows how to build
a clean starter layout. Keeping that logic in `code/` mirrors the habit the
chapter recommends.


```python
from pathlib import Path
import importlib.util
import sys
import tempfile

cwd = Path.cwd().resolve()
BOOK_ROOT = cwd
for candidate in (cwd, cwd.parent, cwd / "4_data"):
    if (candidate / "src").exists() or (candidate / "code").exists():
        BOOK_ROOT = candidate
        break

module_path = BOOK_ROOT / "code" / "project_scaffold.py"
spec = importlib.util.spec_from_file_location("project_scaffold", module_path)
assert spec is not None and spec.loader is not None
project_scaffold = importlib.util.module_from_spec(spec)
spec.loader.exec_module(project_scaffold)
scaffold = project_scaffold.scaffold

print("Book root:", BOOK_ROOT)
```

## Build a Starter Project Tree
The next cell creates a temporary starter project and prints the files it
contains. That makes the recommended layout visible before the delegate writes
anything new.


```python
tmpdir = Path(tempfile.mkdtemp())
starter_root = scaffold(tmpdir / "starter_project")
files = sorted(
    str(path.relative_to(starter_root))
    for path in starter_root.rglob("*")
    if path.is_file()
)
print("\n".join(files))
```

## Compare With the Real Book Layout
The last project cell checks whether the current book already contains the same
kinds of pieces: data, source code, tests, and notebooks.


```python
project_parts = {
    "data": BOOK_ROOT / "data" / "capstone_project_health.csv",
    "src": BOOK_ROOT / "src" / "data_checks.py",
    "tests": BOOK_ROOT / "tests" / "test_data_checks.py",
    "notebooks": BOOK_ROOT / "notebooks" / "01_project_structure.ipynb",
}
lines = []
for name, rel_path in project_parts.items():
    lines.append(
        f"{name:10s} -> {rel_path.relative_to(BOOK_ROOT)} -> "
        f"exists={rel_path.exists()}"
    )
print("\n".join(lines))
```
