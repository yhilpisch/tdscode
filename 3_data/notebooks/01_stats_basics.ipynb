```python
scores = pd.Series(
    [64, 68, 71, 72, 73, 75, 76, 78, 80, 81, 82, 84, 85, 87, 92],
    name='score',
)

print(scores.to_string(index=False))
print()
print(stats_summary.summary(scores).to_string(index=False))
print()
print(
    'median=', round(scores.median(), 1),
    'mean=', round(scores.mean(), 1),
    'std=', round(scores.std(ddof=1), 2),
)
```

This cell compares a balanced and a skewed sample with quartiles and `describe()`.
The contrast shows why distribution shape matters before modeling.

```python
rng = np.random.default_rng(seed=7)
balanced = pd.Series(rng.normal(loc=50, scale=8, size=200), name='balanced')
skewed = pd.Series(rng.gamma(shape=2.0, scale=10.0, size=200), name='skewed')

print('balanced quartiles:')
print(balanced.quantile([0.25, 0.5, 0.75]).to_string())
print()
print('skewed quartiles:')
print(skewed.quantile([0.25, 0.5, 0.75]).to_string())
print()
print('balanced describe:')
print(balanced.describe().round(2).to_string())
print()
print('skewed describe:')
print(skewed.describe().round(2).to_string())
```

This cell compares a balanced and a skewed sample with quartiles and `describe()`.
The contrast shows why distribution shape matters before modeling.

```python
rng = np.random.default_rng(seed=3)
minutes = rng.integers(20, 180, size=60)
noise = rng.normal(loc=0.0, scale=5.0, size=60)
scores = 48 + 0.22 * minutes + noise

X = minutes.reshape(-1, 1)
y = scores

# Split off held-out data before fitting the model.
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=0,
)

model = LinearRegression()
# Fit on the training split.
model.fit(X_train, y_train)

train_pred = model.predict(X_train)
test_pred = model.predict(X_test)

train_mae = np.abs(y_train - train_pred).mean()
test_mae = np.abs(y_test - test_pred).mean()

print('train_mae =', round(float(train_mae), 3))
print('test_mae  =', round(float(test_mae), 3))
```
