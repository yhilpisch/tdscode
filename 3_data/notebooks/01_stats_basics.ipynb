### Goals
- Draw a small sample from a normal distribution.
- Compute mean and standard deviation.


```python
import numpy as np
sample = np.random.normal(loc=0.0, scale=1.0, size=200)
print('mean', sample.mean())
print('std', sample.std(ddof=1))
```

How does the standard deviation change if you double the sample size?
