## Start With NumPy Arrays
We begin with one vector and a simple centered version so whole-array arithmetic
feels concrete before a second signal is introduced.


```python
from pathlib import Path

import numpy as np
import pandas as pd

values = np.array([42, 48, 51, 57, 61, 66])  # monthly signup counts
baseline = values.mean()  # simple reference value
centered = values - baseline  # vectorized arithmetic

print('values:', values.tolist())
print('baseline:', float(baseline))
print('centered:', centered.round(2).tolist())
```

## Combine Two Aligned Signals
The next cell mixes signup counts and study hours. The key habit is that aligned
arrays can be combined directly without writing an explicit loop.


```python
study_hours = np.array([2, 3, 3, 4, 5, 6])  # a second aligned vector
study_score = values * 0.5 + study_hours * 2  # whole-vector calculation

print('study score:', study_score.round(1).tolist())
print('shape:', study_score.shape)
print('dtype:', study_score.dtype)
```

## Move Into a DataFrame
Once the arrays feel natural, we add labels and rows. The printed table shows
how the same information becomes easier to inspect when it has column names.


```python
df = pd.DataFrame({
    'month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'channel': ['web', 'web', 'partner', 'web', 'partner', 'web'],
    'signups': values,
})

print(df)
print()
print(df.describe(include='all'))
```

## Filter to the Rows That Matter
A boolean mask narrows the table to the rows worth attention. That is the first
small version of asking a focused question of a broader dataset.


```python
active = df[df['signups'] >= 55]  # boolean filtering on a column

print('active rows:')
print(active)
```
