```python
from pathlib import Path
import os
import subprocess
import sys

BOOK_NAME = "2_data"
REPO_URL = "https://github.com/yhilpisch/tdscode"

cwd = Path.cwd().resolve()
BOOK_ROOT = None
for candidate in (cwd, *cwd.parents):
    if candidate.name == BOOK_NAME and (candidate / "notebooks").exists():
        BOOK_ROOT = candidate
        break
    if (candidate / BOOK_NAME / "notebooks").exists():
        BOOK_ROOT = candidate / BOOK_NAME
        break

if BOOK_ROOT is None:
    repo_root = Path("/content/tdscode")
    if not repo_root.exists():
        # Clone the companion repo once in Colab.
        subprocess.run(
            ["git", "clone", "--depth", "1", REPO_URL, str(repo_root)],
            check=True,
        )
    BOOK_ROOT = repo_root / BOOK_NAME

os.chdir(BOOK_ROOT)

# Make the book root and the helper folder importable.
for path in (BOOK_ROOT, BOOK_ROOT / "code"):
    if path.exists() and str(path) not in sys.path:
        sys.path.insert(0, str(path))

requirements = BOOK_ROOT / "requirements.txt"
if requirements.exists() and "google.colab" in sys.modules:
    # Keep Colab aligned with the book dependencies.
    subprocess.run(
        [sys.executable, "-m", "pip", "install", "-q", "-r", str(requirements)],
        check=True,
    )

print("Book root:", BOOK_ROOT)
```

```python
from pathlib import Path

import numpy as np
import pandas as pd

values = np.array([42, 48, 51, 57, 61, 66])  # monthly signup counts
baseline = values.mean()  # simple reference value
centered = values - baseline  # vectorized arithmetic

print('values:', values.tolist())
print('baseline:', float(baseline))
print('centered:', centered.round(2).tolist())
```

This cell combines aligned arrays to produce a new score vector. The result shows how
NumPy handles element-wise operations without a loop.

```python
study_hours = np.array([2, 3, 3, 4, 5, 6])  # a second aligned vector
study_score = values * 0.5 + study_hours * 2  # whole-vector calculation

print('study score:', study_score.round(1).tolist())
print('shape:', study_score.shape)
print('dtype:', study_score.dtype)
```

This cell combines aligned arrays to produce a new score vector. The result shows how
NumPy handles element-wise operations without a loop.

```python
df = pd.DataFrame({
    'month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'channel': ['web', 'web', 'partner', 'web', 'partner', 'web'],
    'signups': values,
})

print(df)
print()
print(df.describe(include='all'))
```

This cell filters the table to the higher-signup rows. The output demonstrates how
boolean selection isolates the rows worth attention.

```python
active = df[df['signups'] >= 55]  # boolean filtering on a column

print('active rows:')
print(active)
```
