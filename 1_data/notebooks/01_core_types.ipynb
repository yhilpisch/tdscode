### Why core types matter
This notebook starts with values, not workflows. Beginners need to see how Python
stores information before they can reason about control flow, files, or objects.

### Goals
- Inspect numbers, strings, booleans, lists, tuples, and dictionaries.
- See mutation once, clearly, instead of treating it as magic.
- Save one small artifact that summarizes the example.

## Setup
We make the notebook portable before touching any data or values. That keeps
paths stable and avoids environment-specific confusion later in the book.


```python
from pathlib import Path
import os
import subprocess
import sys

BOOK_NAME = "1_data"
REPO_URL = "https://github.com/yhilpisch/tdscode"

cwd = Path.cwd().resolve()
BOOK_ROOT = None
for candidate in (cwd, *cwd.parents):
    if candidate.name == BOOK_NAME and (candidate / "notebooks").exists():
        BOOK_ROOT = candidate
        break
    if (candidate / BOOK_NAME / "notebooks").exists():
        BOOK_ROOT = candidate / BOOK_NAME
        break

if BOOK_ROOT is None:
    repo_root = Path("/content/tdscode")
    if not repo_root.exists():
        # Clone the companion repo once in Colab.
        subprocess.run(
            ["git", "clone", "--depth", "1", REPO_URL, str(repo_root)],
            check=True,
        )
    BOOK_ROOT = repo_root / BOOK_NAME

os.chdir(BOOK_ROOT)

# Make the book root and the helper folder importable.
for path in (BOOK_ROOT, BOOK_ROOT / "code"):
    if path.exists() and str(path) not in sys.path:
        sys.path.insert(0, str(path))

requirements = BOOK_ROOT / "requirements.txt"
if requirements.exists() and "google.colab" in sys.modules:
    # Keep Colab aligned with the book dependencies.
    subprocess.run(
        [sys.executable, "-m", "pip", "install", "-q", "-r", str(requirements)],
        check=True,
    )

print("Book root:", BOOK_ROOT)
```

## Inspect Core Types
We begin with a compact set of values so the delegate can see numbers, strings,
lists, tuples, booleans, and dictionaries side by side before any abstraction
gets in the way.


```python
price = 19.5
name = 'Ada'
scores = [88, 92, 95]
tags = ('delegate', 'beginner', 'python')
active = True
profile = {'name': name, 'price': price, 'scores': scores, 'tags': tags}

print('Values and types')
for label, value in profile.items():
    print(f'  {label:>8}: {value!r} ({type(value).__name__})')
```

## Mutate a List and Refresh the Summary
A short in-place update makes mutability visible. The next cell changes one
value and immediately recomputes a summary so the consequence is concrete.


```python
scores.append(99)
average = sum(scores) / len(scores)

print(f'Updated scores: {scores}')
print(f'Average score: {average:.1f}')
print(f'Profile keys: {list(profile.keys())}')
```

## Save a Small Artifact
The notebook ends by writing a short summary file. That keeps the lesson tied to
something visible on disk instead of disappearing with the notebook output.


```python
from pathlib import Path

artifact_path = BOOK_ROOT / 'artifacts' / 'core_types_summary.txt'
artifact_path.parent.mkdir(parents=True, exist_ok=True)
artifact_path.write_text(
    'Core types summary\n'
    f'Name: {name}\n'
    f'Price: {price}\n'
    f'Average score: {average:.1f}\n'
    f'Tags: {tags}\n'
    f'Active: {active}\n',
    encoding='utf-8',
)
print(artifact_path.resolve())
print(artifact_path.read_text(encoding='utf-8'))
```
