```python
def plain_loop_completed(tasks: list[str], done: list[bool]) -> list[str]:
    completed = []
    for index in range(len(tasks)):
        if done[index]:
            completed.append(tasks[index])
    return completed

def pythonic_completed(tasks: list[str], done: list[bool]) -> list[str]:
    return [task for task, flag in zip(tasks, done) if flag]

def scorecard(rows: list[tuple[str, int]]) -> list[str]:
    lines = []
    for position, (name, score) in enumerate(sorted(rows, key=lambda item: item[1], reverse=True), start=1):
        lines.append(f'{position}. {name} -> {score}')
    return lines

class StudySession:
    def __init__(self, topic: str, minutes: int, focus: str) -> None:
        self.topic = topic
        self.minutes = minutes
        self.focus = focus

    def hours(self) -> float:
        return self.minutes / 60.0

    def summary(self) -> str:
        return f'{self.topic}: {self.minutes} minutes focused on {self.focus}'
```

This cell continues the worked example for the current section.


```python
tasks = ['read chapter', 'run script', 'write note', 'save artifact']
done = [True, False, True, True]
pairs = [('Ada', 93), ('Bea', 84), ('Cal', 91)]
items = ['csv', 'json', 'parquet', 'csv', 'text', 'json']

print('Plain loop:', plain_loop_completed(tasks, done))
print('Pythonic:', pythonic_completed(tasks, done))
print('Scorecard:')
for line in scorecard(pairs):
    print(f'  {line}')
print(f'Any high scores? {any(score >= 90 for _, score in pairs)}')
print(f'All completed? {all(done)}')

session = StudySession('Pythonic thinking', 35, 'readable code')
json_index = items.index("json")
frequencies = {value: items.count(value) for value in sorted(set(items))}

print(f'Session summary: {session.summary()}')
print(f'Linear search for json: {json_index}')
print(f'Frequencies: {frequencies}')
print(f'Selection sort: {sorted(items)}')
```
