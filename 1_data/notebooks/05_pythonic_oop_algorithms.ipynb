## Compare Plain and Pythonic Solutions
We start by writing the plain-loop version first, then a more compact Pythonic
version. The notebook also adds a small class and two hand-written algorithms so
the contrast stays concrete.


```python
def plain_loop_completed(tasks: list[str], done: list[bool]) -> list[str]:
    completed = []
    for index in range(len(tasks)):
        if done[index]:
            completed.append(tasks[index])
    return completed


def pythonic_completed(tasks: list[str], done: list[bool]) -> list[str]:
    return [task for task, flag in zip(tasks, done) if flag]


def find_first_high_score(rows: list[tuple[str, int]], threshold: int) -> tuple[str, int] | None:
    for name, score in rows:
        if score >= threshold:
            return name, score
    return None


def bubble_sort_scores(rows: list[tuple[str, int]]) -> list[tuple[str, int]]:
    items = rows[:]
    for stop in range(len(items) - 1, 0, -1):
        for index in range(stop):
            if items[index][1] < items[index + 1][1]:
                items[index], items[index + 1] = items[index + 1], items[index]
    return items


def scorecard(rows: list[tuple[str, int]]) -> list[str]:
    return [f'{position}. {name} -> {score}' for position, (name, score) in enumerate(rows, start=1)]


class StudySession:
    def __init__(self, topic: str, minutes: int, focus: str) -> None:
        self.topic = topic
        self.minutes = minutes
        self.focus = focus

    def hours(self) -> float:
        return self.minutes / 60.0

    def summary(self) -> str:
        return f'{self.topic}: {self.minutes} minutes focused on {self.focus}'
```

## Run the Helpers on a Small Study Example
Definitions alone are not enough. The next cell feeds a small study example
through each helper so the delegate can compare the outputs side by side.


```python
tasks = ['read chapter', 'clean csv', 'plot results', 'write notes']
done_flags = [True, False, True, True]
study_rows = [('Ada', 91), ('Grace', 84), ('Linus', 77), ('Sam', 88)]
session = StudySession('Feature engineering', 95, 'careful preprocessing')

plain_completed = plain_loop_completed(tasks, done_flags)
pythonic_list = pythonic_completed(tasks, done_flags)
ranked_rows = bubble_sort_scores(study_rows)
first_high = find_first_high_score(ranked_rows, 85)

print('plain loop:', plain_completed)
print('pythonic:', pythonic_list)
print()
print('ranked scores:')
for line in scorecard(ranked_rows):
    print(' ', line)
print()
print('first high score:', first_high)
print('session:', session.summary())
print('hours:', round(session.hours(), 2))
```

## Save a Ranked Summary
The final artifact keeps the notebook from ending in transient output. It saves
the ranked results and one object summary as a small text file.


```python
artifact_path = BOOK_ROOT / 'artifacts' / 'pythonic_oop_algorithms_summary.txt'
artifact_path.parent.mkdir(parents=True, exist_ok=True)
artifact_path.write_text(
    'Pythonic thinking, OOP, and algorithms summary\n'
    f'Plain completed: {plain_completed}\n'
    f'Pythonic completed: {pythonic_list}\n'
    f'First high score: {first_high}\n'
    f'Session: {session.summary()}\n'
    'Ranked scores:\n' + '\n'.join(scorecard(ranked_rows)) + '\n',
    encoding='utf-8',
)
print(artifact_path.resolve())
print(artifact_path.read_text(encoding='utf-8'))
```
