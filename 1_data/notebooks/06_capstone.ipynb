### Capstone frame
This notebook turns the module into a tiny but complete project. The point is to show
that a beginner can frame a question, keep helper logic tidy, and save a clear
artifact with a README-style explanation.

### Goals
- Sketch a small project structure.
- Write a capstone checklist in plain language.
- Save both the scaffold and a cleaned text artifact.

## Setup
We start by making the notebook portable so the capstone scaffold lands in a
predictable folder and the later paths remain easy to explain.


```python
from pathlib import Path
import os
import subprocess
import sys

BOOK_NAME = "1_data"
REPO_URL = "https://github.com/yhilpisch/tdscode"

cwd = Path.cwd().resolve()
BOOK_ROOT = None
for candidate in (cwd, *cwd.parents):
    if candidate.name == BOOK_NAME and (candidate / "notebooks").exists():
        BOOK_ROOT = candidate
        break
    if (candidate / BOOK_NAME / "notebooks").exists():
        BOOK_ROOT = candidate / BOOK_NAME
        break

if BOOK_ROOT is None:
    repo_root = Path("/content/tdscode")
    if not repo_root.exists():
        # Clone the companion repo once in Colab.
        subprocess.run(
            ["git", "clone", "--depth", "1", REPO_URL, str(repo_root)],
            check=True,
        )
    BOOK_ROOT = repo_root / BOOK_NAME

os.chdir(BOOK_ROOT)

# Make the book root and the helper folder importable.
for path in (BOOK_ROOT, BOOK_ROOT / "code"):
    if path.exists() and str(path) not in sys.path:
        sys.path.insert(0, str(path))

requirements = BOOK_ROOT / "requirements.txt"
if requirements.exists() and "google.colab" in sys.modules:
    # Keep Colab aligned with the book dependencies.
    subprocess.run(
        [sys.executable, "-m", "pip", "install", "-q", "-r", str(requirements)],
        check=True,
    )

print("Book root:", BOOK_ROOT)
```

## Build a Capstone Folder Skeleton
The first capstone step is structural: create a small project tree and write a
README-style note that explains the deliverables in plain language.


```python
from pathlib import Path

root = BOOK_ROOT / 'artifacts' / 'module1_capstone'
for rel in ('notebooks', 'code', 'data', 'notes'):
    (root / rel).mkdir(parents=True, exist_ok=True)

readme = root / 'README.md'
readme.write_text(
    '# Module 1 Capstone Scaffold\n\n'
    'Question: how can I turn a small text-heavy problem into a clean project?\n\n'
    'Deliverables:\n'
    '- one notebook with a clear title, explanation, code, and reflection\n'
    '- one helper script for reusable text/file work\n'
    '- one short README that explains the project and how to run it\n'
    '- one saved text artifact that shows the result of the workflow\n',
    encoding='utf-8',
)

checklist = root / 'notes' / 'capstone_checklist.txt'
checklist.write_text(
    'Define the question in one sentence.\n'
    'Pick a small dataset or a small text file that you can explain.\n'
    'Keep helper code in a separate script if the notebook gets long.\n'
    'Save at least one artifact in a predictable folder.\n'
    'Write a short reflection on what you learned and what to improve.\n',
    encoding='utf-8',
)

print(readme.resolve())
print(checklist.resolve())
```

## Clean and Save a Tiny Text Artifact
The second step adds one concrete output. A few messy lines are cleaned and
saved so the scaffold already contains evidence of work.


```python
notes = [
    'Messy line one with extra spaces   ',
    'Second line that needs cleaning',
    '   third line should become tidy   ',
]
cleaned = [' '.join(line.strip().split()).lower() for line in notes]
cleaned_path = root / 'data' / 'cleaned_notes.txt'
cleaned_path.write_text('\n'.join(cleaned) + '\n', encoding='utf-8')
print(cleaned_path.resolve())
print(cleaned_path.read_text(encoding='utf-8'))
```
