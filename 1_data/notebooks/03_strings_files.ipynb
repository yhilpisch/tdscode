### Why strings and files matter
Text handling is one of the first practical skills in data work. Cleaning a line of
text and saving the result are both small, but they already create real artifacts a
delegate can inspect later.

### Goals
- Clean raw strings with `strip`, `split`, and case normalization.
- Build a file-safe slug from plain text.
- Save a short note to disk and read it back.

```python
from pathlib import Path
import os
import subprocess
import sys

BOOK_NAME = "1_data"
REPO_URL = "https://github.com/yhilpisch/tdscode"

cwd = Path.cwd().resolve()
BOOK_ROOT = None
for candidate in (cwd, *cwd.parents):
    if candidate.name == BOOK_NAME and (candidate / "notebooks").exists():
        BOOK_ROOT = candidate
        break
    if (candidate / BOOK_NAME / "notebooks").exists():
        BOOK_ROOT = candidate / BOOK_NAME
        break

if BOOK_ROOT is None:
    repo_root = Path("/content/tdscode")
    if not repo_root.exists():
        # Clone the companion repo once in Colab.
        subprocess.run(
            ["git", "clone", "--depth", "1", REPO_URL, str(repo_root)],
            check=True,
        )
    BOOK_ROOT = repo_root / BOOK_NAME

os.chdir(BOOK_ROOT)

# Make the book root and the helper folder importable.
for path in (BOOK_ROOT, BOOK_ROOT / "code"):
    if path.exists() and str(path) not in sys.path:
        sys.path.insert(0, str(path))

requirements = BOOK_ROOT / "requirements.txt"
if requirements.exists() and "google.colab" in sys.modules:
    # Keep Colab aligned with the book dependencies.
    subprocess.run(
        [sys.executable, "-m", "pip", "install", "-q", "-r", str(requirements)],
        check=True,
    )

print("Book root:", BOOK_ROOT)
```

```python
def normalize_line(line: str) -> str:
    return ' '.join(line.strip().split()).lower()

def make_slug(text: str) -> str:
    return '_'.join(normalize_line(text).replace(',', '').split())

raw_lines = [
    '  Hello Files  ',
    'Working with Paths  ',
    '  Paths and text should stay readable. ',
]

for line in raw_lines:
    print(f'{line!r} -> {normalize_line(line)!r}')
print(f"Slug example: {make_slug('Module 1 Notes, Draft')}")
```

This cell defines small text-cleaning helpers and applies them to messy strings. The
result shows why trimming, lowercasing, and slug creation matter for file work.

```python
from pathlib import Path

artifact_path = BOOK_ROOT / 'artifacts' / 'strings_files_note.txt'
artifact_path.parent.mkdir(parents=True, exist_ok=True)
cleaned = [normalize_line(line) for line in raw_lines]
artifact_path.write_text('\n'.join(cleaned) + '\n', encoding='utf-8')

print(artifact_path.resolve())
print('Saved lines:')
for line in artifact_path.read_text(encoding='utf-8').splitlines():
    print(f'  {line}')
```
