### Why this notebook exists
Control flow gives shape to a program. Functions then package a repeated idea so you
can reuse it without copying the same logic everywhere.

- Use a `while` loop to show visible state.

## Setup
We make the notebook portable first so the same import and artifact paths work
everywhere the delegate opens it.


```python
from pathlib import Path
import os
import subprocess
import sys

BOOK_NAME = "1_data"
REPO_URL = "https://github.com/yhilpisch/tdscode"

cwd = Path.cwd().resolve()
BOOK_ROOT = None
for candidate in (cwd, *cwd.parents):
    if candidate.name == BOOK_NAME and (candidate / "notebooks").exists():
        BOOK_ROOT = candidate
        break
    if (candidate / BOOK_NAME / "notebooks").exists():
        BOOK_ROOT = candidate / BOOK_NAME
        break

if BOOK_ROOT is None:
    repo_root = Path("/content/tdscode")
    if not repo_root.exists():
        # Clone the companion repo once in Colab.
        subprocess.run(
            ["git", "clone", "--depth", "1", REPO_URL, str(repo_root)],
            check=True,
        )
    BOOK_ROOT = repo_root / BOOK_NAME

os.chdir(BOOK_ROOT)

# Make the book root and the helper folder importable.
for path in (BOOK_ROOT, BOOK_ROOT / "code"):
    if path.exists() and str(path) not in sys.path:
        sys.path.insert(0, str(path))

requirements = BOOK_ROOT / "requirements.txt"
if requirements.exists() and "google.colab" in sys.modules:
    # Keep Colab aligned with the book dependencies.
    subprocess.run(
        [sys.executable, "-m", "pip", "install", "-q", "-r", str(requirements)],
        check=True,
    )

print("Book root:", BOOK_ROOT)
```

## Define Reusable Flow Helpers
The next cell turns branching and looping into named helpers. That is the first
step from one-off code toward code that can be reused and tested.


```python
def score_band(score: int) -> str:
    if score >= 90:
        return 'excellent'
    if score >= 80:
        return 'strong'
    if score >= 70:
        return 'solid'
    if score >= 60:
        return 'passing'
    return 'needs review'

def countdown(start: int = 3) -> list[int]:
    numbers = []
    current = start
    while current >= 0:
        numbers.append(current)
        current -= 1
    return numbers
```

## Run the Helpers on Example Inputs
Once the functions exist, we feed them visible examples. The output matters
because it shows what the helper actually does, not just how it is written.


```python
scores = [58, 74, 81, 93]
words = ['python', 'data', 'analysis', 'loops', 'functions']

print('Score bands')
for score in scores:
    print(f'  {score} -> {score_band(score)}')

long_words = [word for word in words if len(word) >= 5]
print(f'Long words: {long_words}')
print(f'Countdown: {countdown(3)}')
```

## Save the Results
We close by writing a small text summary to disk. Even a tiny artifact makes the
workflow feel more like a project and less like a disposable scratchpad.


```python
from pathlib import Path

artifact_path = BOOK_ROOT / 'artifacts' / 'flow_functions_summary.txt'
artifact_path.parent.mkdir(parents=True, exist_ok=True)
artifact_path.write_text(
    'Flow and functions summary\n'
    f'Scores: {scores}\n'
    f'Countdown: {countdown(3)}\n'
    f'Long words: {long_words}\n',
    encoding='utf-8',
)
print(artifact_path.resolve())
print(artifact_path.read_text(encoding='utf-8'))
```
