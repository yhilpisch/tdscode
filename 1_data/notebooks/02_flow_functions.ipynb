### Why this notebook exists
Control flow gives shape to a program. Functions then package a repeated
idea so you can reuse it without copying the same logic everywhere.

### Goals
- Use `if` and `elif` to classify values.
- Use a `while` loop to show visible state.
- Turn repeated logic into small functions.

```python
def score_band(score: int) -> str:
    if score >= 90:
        return 'excellent'
    if score >= 80:
        return 'strong'
    if score >= 70:
        return 'solid'
    if score >= 60:
        return 'passing'
    return 'needs review'

def countdown(start: int = 3) -> list[int]:
    numbers = []
    current = start
    while current >= 0:
        numbers.append(current)
        current -= 1
    return numbers
```

```python
scores = [58, 74, 81, 93]
words = ['python', 'data', 'analysis', 'loops', 'functions']

print('Score bands')
for score in scores:
    print(f'  {score} -> {score_band(score)}')

long_words = [word for word in words if len(word) >= 5]
print(f'Long words: {long_words}')
print(f'Countdown: {countdown(3)}')
```

```python
from pathlib import Path

artifact_path = Path('artifacts') / 'flow_functions_summary.txt'
artifact_path.parent.mkdir(parents=True, exist_ok=True)
artifact_path.write_text(
    'Flow and functions summary\n'
    f'Scores: {scores}\n'
    f'Countdown: {countdown(3)}\n'
    f'Long words: {long_words}\n',
    encoding='utf-8',
)
print(artifact_path.resolve())
print(artifact_path.read_text(encoding='utf-8'))
```
