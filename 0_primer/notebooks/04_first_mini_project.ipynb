### Project frame
This notebook closes the primer with a complete but very small artifact. The
point is not technical depth. The point is to finish a loop cleanly: explain the
purpose, run some code, save a result, and write down what the notebook proves.

### Goals
- Build one tiny notebook-centered artifact.
- Save both a text summary and a plot inside the project.
- End with a short reflection that could become a portfolio note.

This cell continues the worked example for the current section.


This cell prepares the notebook for local or Colab execution. It finds the book root, clones the companion repo when needed, and makes the working directory predictable.


```python
from pathlib import Path
import os
import subprocess
import sys

BOOK_NAME = "0_primer"
REPO_URL = "https://github.com/yhilpisch/tdscode"

cwd = Path.cwd().resolve()
BOOK_ROOT = None
for candidate in [cwd, *cwd.parents]:
    if candidate.name == BOOK_NAME and (candidate / "notebooks").exists():
        BOOK_ROOT = candidate
        break
    if (candidate / BOOK_NAME / "notebooks").exists():
        BOOK_ROOT = candidate / BOOK_NAME
        break

if BOOK_ROOT is None:
    repo_root = Path("/content/tdscode")
    if not repo_root.exists():
        subprocess.run(["git", "clone", "--depth", "1", REPO_URL, str(repo_root)], check=True)
    BOOK_ROOT = repo_root / BOOK_NAME

os.chdir(BOOK_ROOT)
if str(BOOK_ROOT) not in sys.path:
    sys.path.insert(0, str(BOOK_ROOT))

code_dir = BOOK_ROOT / "code"
if code_dir.exists() and str(code_dir) not in sys.path:
    sys.path.insert(0, str(code_dir))

requirements = BOOK_ROOT / "requirements.txt"
if requirements.exists() and "google.colab" in sys.modules:
    subprocess.run([sys.executable, "-m", "pip", "install", "-q", "-r", str(requirements)], check=True)

print("Book root:", BOOK_ROOT)
```

```python
from pathlib import Path

name = 'Ada'
study_goal = 'build a reliable learning workflow'
study_hours = [1.0, 1.5, 2.0, 0.5]
average_hours = sum(study_hours) / len(study_hours)

print(f'Name: {name}')
print(f'Study goal: {study_goal}')
print(f'Study hours: {study_hours}')
print(f'Average study hours: {average_hours:.2f}')
```

### Save a text summary
A small summary file makes the notebook easier to revisit because the project
keeps a plain-text explanation alongside the interactive notebook itself.

This cell continues the worked example for the current section.


```python
artifact_dir = BOOK_ROOT / 'artifacts' / 'first_mini_project'
artifact_dir.mkdir(parents=True, exist_ok=True)
summary_path = artifact_dir / 'project_summary.txt'

summary_lines = [
    'Primer Mini Project',
    f'Name: {name}',
    f'Study goal: {study_goal}',
    f'Average study hours: {average_hours:.2f}',
    'This notebook demonstrates Markdown, code, saved output, and reflection.',
]
summary_path.write_text('\n'.join(summary_lines) + '\n', encoding='utf-8')
print(summary_path.resolve())
print(summary_path.read_text(encoding='utf-8'))
```

### Save a small figure
A chart is not required for the primer, but it is a useful final proof that a
notebook can produce a visual artifact and save it into a predictable folder.

This cell continues the worked example for the current section.


```python
figure_path = artifact_dir / 'study_hours.svg'

curve_points = []
for index, hours in enumerate(study_hours, start=1):
    x_position = 40 + (index - 1) * 120
    y_position = 220 - hours * 70
    curve_points.append(f'{x_position},{y_position:.1f}')

svg_text = f'''<svg xmlns="http://www.w3.org/2000/svg" width="420" height="260" viewBox="0 0 420 260">
  <rect width="100%" height="100%" fill="white" />
  <line x1="40" y1="220" x2="380" y2="220" stroke="#999" stroke-width="1" />
  <line x1="40" y1="30" x2="40" y2="220" stroke="#999" stroke-width="1" />
  <polyline fill="none" stroke="#0B3C78" stroke-width="3" points="{' '.join(curve_points)}" />
  <text x="40" y="22" font-size="18" font-family="Arial, sans-serif">Study Hours Across Four Sessions</text>
</svg>'''

figure_path.write_text(svg_text, encoding='utf-8')
print(f'Saved figure to: {figure_path.resolve()}')
```
