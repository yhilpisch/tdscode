### Why start in Colab
Colab removes local setup friction, which is useful at the very beginning of the
program. The real point is not the browser itself. The point is to learn the
study loop in a low-friction setting: read, run, inspect, reflect, and save.

### Goals
- Confirm the runtime and active Python executable.
- Practise the difference between a Markdown explanation and a code cell.
- Save a tiny artifact so the notebook leaves visible evidence of work.
- End with one reflection that could live in a learning journal.

### Create a tiny artifact
Use `Path` so the save location stays explicit and easy to reason about. Even a
small text file is enough to demonstrate the full loop from execution to saved
result.

This cell continues the worked example for the current section.


```python
from pathlib import Path

artifact_dir = Path('artifacts')
artifact_dir.mkdir(parents=True, exist_ok=True)
artifact_path = artifact_dir / 'hello_colab.txt'

lines = [
    'The Data Scientist · Primer',
    'Notebook: 01_colab_quickstart.ipynb',
    'Reflection: I can run code and save the result as a file.',
]
artifact_path.write_text('\n'.join(lines) + '\n', encoding='utf-8')

print(f'Saved artifact to: {artifact_path.resolve()}')
print(artifact_path.read_text(encoding='utf-8'))
```

### Why this tiny step matters
The code above is deliberately modest. It still proves four important things:

- the notebook can run from top to bottom,
- the output is visible in the notebook,
- the work leaves a saved artifact on disk,
- and you can explain the result in plain language.
