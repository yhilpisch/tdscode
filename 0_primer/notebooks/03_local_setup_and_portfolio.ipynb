### Why this notebook exists
Colab is a good start, but durable work needs a local home. This notebook keeps
three ideas connected: local environment awareness, clean relative paths, and a
small project structure that can grow into a portfolio.

### Goals
- Inspect the current Python interpreter.
- Compare a clean project folder with an environment that lives elsewhere.
- Create a tiny portfolio skeleton with a README and a learning-journal note.

## Setup
We begin by making the notebook location explicit. Once the book root is
stable, the rest of the examples can talk about paths and project structure
without hidden assumptions.


```python
from pathlib import Path
import os
import subprocess
import sys

BOOK_NAME = "0_primer"
REPO_URL = "https://github.com/yhilpisch/tdscode"

cwd = Path.cwd().resolve()
BOOK_ROOT = None
for candidate in [cwd, *cwd.parents]:
    if candidate.name == BOOK_NAME and (candidate / "notebooks").exists():
        BOOK_ROOT = candidate
        break
    if (candidate / BOOK_NAME / "notebooks").exists():
        BOOK_ROOT = candidate / BOOK_NAME
        break

if BOOK_ROOT is None:
    repo_root = Path("/content/tdscode")
    if not repo_root.exists():
        subprocess.run(["git", "clone", "--depth", "1", REPO_URL, str(repo_root)], check=True)
    BOOK_ROOT = repo_root / BOOK_NAME

os.chdir(BOOK_ROOT)
if str(BOOK_ROOT) not in sys.path:
    sys.path.insert(0, str(BOOK_ROOT))

code_dir = BOOK_ROOT / "code"
if code_dir.exists() and str(code_dir) not in sys.path:
    sys.path.insert(0, str(code_dir))

requirements = BOOK_ROOT / "requirements.txt"
if requirements.exists() and "google.colab" in sys.modules:
    subprocess.run([sys.executable, "-m", "pip", "install", "-q", "-r", str(requirements)], check=True)

print("Book root:", BOOK_ROOT)
```

### Environment commands by operating system
The exact path can differ, but the reasoning stays the same: keep the project
folder readable and keep the environment location intentional.

The commands below are labeled by operating system so the delegate
can see the difference without mixing macOS/Linux and Windows variants
together.


```python
setup_commands = {
    'macOS/Linux': [
        'python3 -m venv ~/Temp/venv/ds_primer',
        'source ~/Temp/venv/ds_primer/bin/activate',
        'python -m pip install --upgrade pip',
        'which python',
    ],
    'Windows': [
        r'py -m venv C:\Temp\venv\ds_primer',
        r'C:\Temp\venv\ds_primer\Scripts\activate',
        'python -m pip install --upgrade pip',
        'where python',
    ],
}

for label, commands in setup_commands.items():
    print(label)
    for command in commands:
        print(f'  {command}')
```

### Create a tiny portfolio skeleton
This cell writes a minimal structure inside `artifacts/` so the notebook leaves
behind something concrete: a project folder, a README, and a short note about
what to do next.

This cell creates a tiny portfolio-style folder inside
`artifacts/`. The structure is deliberately small, but it already separates
notebooks, code, figures, and notes.


```python
project_root = BOOK_ROOT / 'artifacts' / 'primer_portfolio_notebook'
for rel_path in ['notebooks', 'code', 'figures', 'notes']:
    (project_root / rel_path).mkdir(parents=True, exist_ok=True)

readme_path = project_root / 'README.md'
readme_path.write_text(
    '# Primer Portfolio Notebook\n\n'
    'Purpose: practise a clean beginner project structure.\n\n'
    'Contents:\n'
    '- notebooks/ for interactive work\n'
    '- code/ for helper scripts\n'
    '- figures/ for saved visuals\n'
    '- notes/ for reflections\n',
    encoding='utf-8',
)

journal_path = project_root / 'notes' / 'learning_journal.txt'
journal_path.write_text(
    'Attempted: set up a tiny primer portfolio structure.\n'
    'Worked: relative paths kept everything easy to find.\n'
    'Next: add one small notebook artifact and a first commit.\n',
    encoding='utf-8',
)

print(readme_path.resolve())
print(journal_path.resolve())
```

### What to notice
The notebook did not need an advanced toolchain to create a project skeleton.
It only needed clear path decisions and a small amount of written context.
Those two habits scale surprisingly well.
