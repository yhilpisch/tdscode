### Notebook habits to keep runs reproducible
A notebook is useful only when another person, or your future self, can rerun
it without guesswork. This notebook turns that idea into a short repeatable
checklist.

### Session checklist
- Start with a short Markdown statement of the question.
- Keep imports near the top.
- Use relative paths instead of hunting through the file system manually.
- Save one small artifact before closing the session.
- Write one reflection line about what worked or what needs attention next.

This cell prepares the notebook for local or Colab execution. It finds the book root, clones the companion repo when needed, and makes the working directory predictable.


```python
from pathlib import Path
import os
import subprocess
import sys

BOOK_NAME = "0_primer"
REPO_URL = "https://github.com/yhilpisch/tdscode"

cwd = Path.cwd().resolve()
BOOK_ROOT = None
for candidate in [cwd, *cwd.parents]:
    if candidate.name == BOOK_NAME and (candidate / "notebooks").exists():
        BOOK_ROOT = candidate
        break
    if (candidate / BOOK_NAME / "notebooks").exists():
        BOOK_ROOT = candidate / BOOK_NAME
        break

if BOOK_ROOT is None:
    repo_root = Path("/content/tdscode")
    if not repo_root.exists():
        subprocess.run(["git", "clone", "--depth", "1", REPO_URL, str(repo_root)], check=True)
    BOOK_ROOT = repo_root / BOOK_NAME

os.chdir(BOOK_ROOT)
if str(BOOK_ROOT) not in sys.path:
    sys.path.insert(0, str(BOOK_ROOT))

code_dir = BOOK_ROOT / "code"
if code_dir.exists() and str(code_dir) not in sys.path:
    sys.path.insert(0, str(code_dir))

requirements = BOOK_ROOT / "requirements.txt"
if requirements.exists() and "google.colab" in sys.modules:
    subprocess.run([sys.executable, "-m", "pip", "install", "-q", "-r", str(requirements)], check=True)

print("Book root:", BOOK_ROOT)
```

```python
from pathlib import Path
import json

session_dir = BOOK_ROOT / 'artifacts'
session_dir.mkdir(parents=True, exist_ok=True)
notes_path = session_dir / 'session_notes.json'

session_notes = {
    'question': 'What makes a notebook rerunnable?',
    'habits': [
        'Keep code in a logical top-to-bottom order.',
        'Avoid hidden state by restarting and rerunning the notebook.',
        'Save one artifact so the session leaves evidence behind.',
    ],
    'next_step': 'Move from notebook habits to a local project structure.',
}

notes_path.write_text(json.dumps(session_notes, indent=2) + '\n', encoding='utf-8')
print(notes_path.resolve())
print(notes_path.read_text(encoding='utf-8'))
```

### Inspect the saved result
The output above should be readable without opening any extra tool. That is a
useful standard for beginner work: save something small, then immediately check
that it contains the information you thought you wrote.
